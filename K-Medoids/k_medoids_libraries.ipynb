```python
df = pd.read_csv('Iris.csv')

print(df.head())
print(df.info())
df
```

```python
# removes no numeric columns 
X = df.iloc[: ,:-1]
X
```

```python
# normalize the data with StandardScaler 
scaler = StandardScaler()
X = scaler.fit_transform(X)
X.shape #shows (cols, rows)
```

```python
k =3 # there are three catagories of flowers

np.random.seed(random_seed)
medoid_indices = np.random.choice(X.shape[0], size=k, replace= False)
medoids = X[medoid_indices]

print("medoid indices: ",medoid_indices)
print("\nmediods: ", medoids)
```

```python
def pairwise_distances(X,medoids):
    """
    Compute Elucidian distances between two set of points 
    A: shape (n_samples, n_features)
    B: shape (m_samples, n_features)
    returns: distances shape (n_samples, m_samples)
    """
    X_sq = np.sum(X**2, axis=1, keepdims=True)
    medoids_sq = np.sum(medoids**2, axis=1, keepdims=True)
    cross = X @ medoids.T # X * medoids.T
    print(cross.shape)
    dists = np.sqrt(X_sq + medoids_sq - (2*cross))
    return dists
```

```python
distances = pairwise_distances(X, medoids)
labels = np.argmin(distances , axis=1)
```
